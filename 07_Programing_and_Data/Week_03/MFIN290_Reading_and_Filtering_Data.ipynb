```python
# Fix datadate
df = pd.read_csv('fina.csv', parse_dates=['datadate'], date_format='%Y%m%d')
df.head(20)
```

```python
# After fixing dates, we can do a lot of useful things like...
import datetime as dt
df['year'] = df['datadate'].dt.year
df.head()
```

```python
# Let's just sort data so that it looks neat
df.sort_values(['conm','year'], inplace=True)
df.head(20)
```

```python
df.info()
```

### Filtering 
aka 'Conditional selection'




```python
# It creates 3502 boolean across observations
df['ni'] > 0
```

```python
# keeps only rows where the condition is true
df[df['ni']>0]
```

```python
# doing the same thing with a different code
df.loc[df['ni']>0]
```

```python
#the above is not "inplace".  If you want to keep the subset, generate a new DataFrame:
df2 = df[df['ni']>0]
df2
```

```python
# If you want to only keep tic
df2 = df[df['ni']>0][['tic']]
df2
```

```python
# If you want to keep tic datadate ni sale
df2 = df[df['ni']>0][['tic','datadate','ni','sale']]
df2
```

```python
#use & and | for multiple conditions
df2 = df[(df['ni']>0) & (df['lt']>0)]
df2
```

```python
# .isin() creates a boolean series whenter a condition is met.  Create the boolean, then subset the data by it.
fang = df['tic'].isin(['FB','AAPL','NFLX','GOOGL'])
#fang
df3 = df[fang]
df3.head(8)
```

```python
# Another condition
filteryear = df['year'].isin([2017, 2018])
filteryear
```

```python
df3 = df[fang & filteryear]
df3
```
